**PLOTTING SIGMOID FUNCTION**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

$S_{w,b}(x)=\frac{1}{1+e^{-(wx+b))}}$

```python
def sigmoid(x,w,b):
  return 1/(1+np.exp(-(w*x+b)))
```

```python
sigmoid(1,0.5,0)
```

```python
W=0.4  # @param {type: "slider", min:-2, max: 2,steps:0.1}
b=0   # @param {type: "slider", min: -2, max: 2,steps:0.1}
X=np.linspace(-10,10,100)
Y=sigmoid(X,W,b)

plt.plot(X,Y)
plt.show()
```

**PLOTTING A 3 DIMENSIONAL SIGMOID FUNCTION**

```python
def sigmoid_2d(x1,x2,w1,w2,b):
  return 1/(1+np.exp(-(w1*x1+w2*x2+b)))
```

```python
from mpl_toolkits import mplot3d
```

```python
X1=np.linspace(-10,10,100)
X2=np.linspace(-10,10,100)

XX1,XX2=np.meshgrid(X1,X2)

print(X1.shape,X2.shape,XX1.shape,XX2.shape)
```

```python
w1=0.5
w2=0.5
b=0
Y=sigmoid_2d(XX1,XX2,w1,w2,b)
print(Y.shape)
```

```python
my_cmap=matplotlib.colors.LinearSegmentedColormap.from_list("custom_cmap", ["red", "violet", "blue"])
```

```python
plt.contourf(XX1,XX2,Y,cmap=my_cmap)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(XX1,XX2,Y,cmap="viridis")
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")

# Adjust the viewing angle

plt.show()
```

**COMPUTE LOSS FOR A GIVEN DATASET**

```python
w_unknown = 0.5
b_unknown = 0.25
X=np.random.random(10) *20-10
Y=sigmoid(X,w_unknown,b_unknown)
```

```python
plt.plot(X,Y,"*")
plt.show()
```

```python
def calculate_loss(X,Y,w_est,b_est):
  loss=0
  for x,y in zip(X,Y):
    loss+=0.5*(y-sigmoid(x,w_est,b_est))**2
  return loss
```

```python
W=np.linspace(0,1,100)
B=np.linspace(-1,1,100)
WW,BB=np.meshgrid(W,B)
loss=np.zeros(WW.shape)
```

```python
WW.shape
```

```python
for i in range(WW.shape[0]):
  for j in range(WW.shape[1]):
    loss[i,j]=calculate_loss(X,Y,WW[i,j],BB[i,j])
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(WW,BB,loss,cmap="viridis")
ax.set_xlabel("W")
ax.set_ylabel("b")
ax.set_zlabel("Loss")

# Adjust the viewing angle

plt.show()
```
